```python
import pandas as pd
import os

# Define file paths
input_file = "/Users/chsmac/Downloads/Traffic_Crashes_Crashes1.csv"
output_file = "/Users/chsmac/Downloads/Reduced_Traffic_Crashes_Crashes.csv"

# Load the dataset
df = pd.read_csv(input_file)

# Display original file size
original_size = os.path.getsize(input_file) / (1024 * 1024)  # Convert bytes to MB
print(f"Original File Size: {original_size:.2f} MB")

# Keep only essential columns (modify as needed)
columns_to_keep = ['CRASH_DATE', 'LATITUDE', 'LONGITUDE', 'INJURIES_TOTAL', 'CRASH_TYPE']
df = df[columns_to_keep]

# Convert to more efficient data types
df['LATITUDE'] = df['LATITUDE'].astype('float32')
df['LONGITUDE'] = df['LONGITUDE'].astype('float32')
df['INJURIES_TOTAL'] = pd.to_numeric(df['INJURIES_TOTAL'], errors='coerce').fillna(0).astype('int8')

# Convert date column to string (saves space)
df['CRASH_DATE'] = pd.to_datetime(df['CRASH_DATE']).dt.strftime('%Y-%m-%d')

# Save reduced file and check new size
df.to_csv(output_file, index=False)

reduced_size = os.path.getsize(output_file) / (1024 * 1024)  # Convert bytes to MB
print(f"Reduced File Size: {reduced_size:.2f} MB")

# If file is still >100MB, sample rows
if reduced_size > 100:
    fraction = 100 / reduced_size  # Calculate fraction to sample
    df_sampled = df.sample(frac=fraction, random_state=42)
    df_sampled.to_csv(output_file, index=False)

    final_size = os.path.getsize(output_file) / (1024 * 1024)  # Convert bytes to MB
    print(f"Final File Size After Sampling: {final_size:.2f} MB")

print(f"Optimized file saved at: {output_file}")
```
